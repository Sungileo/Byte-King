```python
from create_heart import get_heart_boundary_points, generate_heart_coordinates
# from create_bottom_heart import *
from create_sharp import *

from shapely import wkt
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd

# 서울시청(37.566826, 126.978656)을 중심으로 2km 크기의 하트 모양 생성
# center_lat = 37.566826
# center_lon = 126.978656

# 서울 남부
# center_lat, center_lon = 37.484208,126.929676

# 상수역
center_lat, center_lon = 37.5554, 126.9214
# points = get_heart_boundary_points(center_lat, center_lon, size_km=2.5, num_points=50)
points = get_heart_boundary_points_from_bottom(center_lat, center_lon, size_km=0.7, num_points=20)

# 신림역, 보라매공원
pointdata = pd.DataFrame([{"name":f"point_{i}", "geometry":Point(lon, lat)} for i,(lat, lon) in enumerate(points)])

# pointdata['geometry'] = pointdata['geometry'].apply(wkt.loads)
pointdata = gpd.GeoDataFrame(pointdata, geometry='geometry', crs="epsg:4326")
```

```python
pointdata.head()
```

```python
import folium
from folium import Circle

M = folium.Map(location=[pointdata.geometry.y.mean(), pointdata.geometry.x.mean()], zoom_start=13, tiles='cartodbpositron')

for idx, row in pointdata.iterrows():
    Circle(location = [row.geometry.y, row.geometry.x],
           radius=40,
           tooltip=row["name"],
           color = "red").add_to(M)
    
M
```

```python
from find_best_line import find_optimized_path
from tqdm import tqdm

ans = pd.DataFrame()

for i in tqdm(range(len(points))):
    lon1,lat1 = points[i-1]
    lon2,lat2 = points[i]
    
    start = Point(lat1, lon1)
    end = Point(lat2, lon2)

    res = find_optimized_path(start, end)

    ans = pd.concat([ans, res], axis = 0)

ans.drop_duplicates(inplace=True)
```

```python
ans['링크 길이'].sum()
```

```python
M = folium.Map(location=[37.484208,126.929676], zoom_start=17)

# GeoDataFrame의 각 LINESTRING을 지도에 추가
for idx, row in ans.iterrows():
    # LINESTRING 좌표 추출
    coordinates = [(y, x) for x, y in row.geometry.coords]

    # LineString 추가
    folium.PolyLine(
        locations=coordinates,
        weight=5,
        color='red',
        opacity=1.0,
        popup=f"LineString ID: {idx}"  # 팝업 정보 (필요에 따라 수정)
    ).add_to(M)

M
```
