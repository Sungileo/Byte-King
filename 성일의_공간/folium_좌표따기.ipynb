```python
import folium

# 서울 중심 좌표
seoul_center = [37.5665, 126.9780]

# 지도 생성
m = folium.Map(
    location=seoul_center,
    zoom_start=11,
    width=600,
    height=900
)

# 클릭 이벤트를 처리할 JavaScript 추가
m.get_root().html.add_child(folium.Element("""
    <div id="coordinates" style="
        position: fixed;
        top: 20px;
        right: 20px;
        background: white;
        padding: 15px;
        border-radius: 8px;
        box-shadow: 0 0 10px rgba(0,0,0,0.1);
        z-index: 1000;
        font-family: Arial, sans-serif;">
        <div style="margin: 10px 0;">
            <span style="font-weight: bold; width: 60px; display: inline-block;">위도:</span>
            <span id="lat" style="color: #0066cc; font-family: monospace; font-size: 18px;">-</span>
        </div>
        <div style="margin: 10px 0;">
            <span style="font-weight: bold; width: 60px; display: inline-block;">경도:</span>
            <span id="lng" style="color: #0066cc; font-family: monospace; font-size: 18px;">-</span>
        </div>
    </div>
    
    <script>
        map.on('click', function(e) {
            var lat = e.latlng.lat.toFixed(4);
            var lng = e.latlng.lng.toFixed(4);
            
            // 위경도 표시 업데이트
            document.getElementById('lat').textContent = lat;
            document.getElementById('lng').textContent = lng;
            
            // 마커 추가
            L.marker([lat, lng]).addTo(map)
                .bindPopup('위도: ' + lat + '<br>경도: ' + lng)
                .openPopup();
            
            // 콘솔에도 출력
            console.log('위도: ' + lat + ', 경도: ' + lng);
        });
    </script>
"""))

# HTML 파일로 저장
m.save('map.html')

print("map.html 파일이 생성되었습니다. 웹 브라우저로 열어주세요!")
```
